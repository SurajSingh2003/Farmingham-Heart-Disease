```python
## Loading Libraries packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

## loading sklearn packeges required for the analysis

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df= pd.read_csv('framingham.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.drop(['education'], axis=1, inplace=True)
df.drop(['male'], axis=1, inplace=True)
```

```python
df.head()
```

```python
pd.options.display.float_format='{:.2f}'.format
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(),annot = True)
plt.show()
```

```python
df.isnull().sum()
```

```python
df.hist('cigsPerDay')
df.hist('BPMeds')
df.hist('totChol')
df.hist('glucose')
df.hist('heartRate')
df.hist('BMI')
```

```python
df['totChol']=df['totChol'].fillna(df.totChol.median())
```

```python
df['glucose']=df['glucose'].fillna(df.glucose.median())
df['cigsPerDay']=df['cigsPerDay'].fillna(df.cigsPerDay.median())
df['BPMeds']=df['BPMeds'].fillna(df.BPMeds.median())
df['heartRate']=df['heartRate'].fillna(df.heartRate.median())
df['BMI']=df['BMI'].fillna(df.BMI.median())
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
x = df.iloc[:,0:13].values
y=df['TenYearCHD']
```

```python
x.shape
```

```python
print(x[0:6,:])
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x[:,0:13]=sc.fit_transform(x[:,0:13])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lm= LogisticRegression()
lm.fit(x_train,y_train)
```

```python
print(lm.coef_)
print(lm.intercept_)
```

```python
display(lm.predict_proba(x_test)[:10])
display(lm.predict(x_test)[:10])
```

```python
from sklearn.metrics import accuracy_score
print("Logistic testing accuracy score is %2.2f" %accuracy_score(y_test,lm.predict(x_test)))
print("Logistic training accuracy score is %2.2f" %accuracy_score(y_train,lm.predict(x_train)))
```

```python
from sklearn.metrics import roc_auc_score

# Let us measure the Logistic Regression model by the help of ROC and AUC curve

logistic_roc_auc = roc_auc_score(y_test,lm.predict(x_test))
print(logistic_roc_auc)
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test,lm.predict(x_test))
```

```python
plt.figure()
plt.plot(fpr, tpr, label = 'ROC Curve (area= %0.2f)' %logistic_roc_auc)
plt.plot([0,1], [0,1], 'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Exampls')
plt.legend(loc='lower right')
plt.show()


print('Logistic Auc = %2.2f'%logistic_roc_auc)
```

```python
y_pred= lm.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
